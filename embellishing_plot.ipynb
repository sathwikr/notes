```python
#expression, latex style
pl.plot(x, -x*x+4*x-5, 'r', linewidth =2)
pl.title(r'Parabolic function $-x^2+4x+5$')
pl.xlabel(r'$x$')
pl.ylabel(r'$y$')
pl.annotate('local maxima', xy = (2, -1)) #1st arg - name, 2-coordinate
pl.xlim()
pl.ylim()
```

```python
#expression, latex style
pl.plot(x, -x*x+4*x-5, 'r', linewidth =2)
pl.title(r'Parabolic function $-x^2+4x+5$')
pl.xlabel(r'$x$')
pl.ylabel(r'$y$')
pl.annotate('local maxima', xy = (2, -1)) #1st arg - name, 2-coordinate
pl.ylim(-15, 2)
pl.xlim(-4, 5)
```

```python
#Make annontation called root(-4, 0)
#expression, latex style
pl.plot(x, -x*x+4*x-5, 'r', linewidth =2)
pl.title(r'Parabolic function $-x^2+4x+5$')
pl.xlabel(r'$x$')
pl.ylabel(r'$y$')
pl.annotate('root', xy = (-4,0))
pl.annotate('local maxima', xy = (2, -1)) #1st arg - name, 2-coordinate
pl.ylim(-15, 2)
pl.xlim(-4, 5)
```
